```python
from nltk.corpus import wordnet as wn
```

## Print all noun words present in wordnet

```python
for synset in wn.all_synsets('n') :
    print(synset.name().split('.')[0])
```

## Print the different senses or meanings of the given word 

```python
word = input()
synsets = wn.synsets(word)
for ss in synsets :
    print(ss.name().split('.')[0].upper(),ss.definition())
```

## Print all synonyms for the given word. Also count them

```python
word = input()
synonyms = []
for ss in wn.synsets(word) :
    for lemma in ss.lemmas() :
        synonyms.append(lemma.name())
print("No. of synonyms:", len(synonyms), synonyms)
```

## Print the antonyms of the given word

```python
word = input()
antonyms = []
for ss in wn.synsets(word) :
    for lemma in ss.lemmas() :
        if lemma.antonyms(): 
            antonyms.append(lemma.antonyms()[0].name()) 
print("No. of antonyms:", len(antonyms), antonyms)
```

## Print the number of senses of the given word in all parts-of-speech category like noun, verb, adverb and adjective

```python
word = input()
noun = verb = adverb = adjective = 0
tags = []
for ss in wn.synsets(word) :
    tags.append(ss.name().split('.')[1])
noun = tags.count('n')
verb = tags.count('v')
adverb = tags.count('r')
adjective = tags.count('a')
print("No of nouns = {}, No of verbs = {}, No of adverbs = {}, No of adjectives = {}".format(noun, verb, adverb, adjective))
```

## Find the semantic similarity between the given words

```python
word1 = input()
word2 = input()
word1 = wn.synsets(word1)[0].name()
word2 = wn.synsets(word2)[0].name()
word1 = wn.synset(word1)
word2 = wn.synset(word2)
print(word1.path_similarity(word2))
```

## For the given word print its hypernym and co-hyponyms 

```python
word = input()
hypernyms = []
synsets = wn.synsets(word)
for ss in synsets :
    hypernyms.extend(ss.hypernyms())
print("Hypernyms:", hypernyms)
cohypernyms = []
for hyp in hypernyms :
    cohypernyms.extend(hyp.hypernyms())
print("Co-hypernyms:", cohypernyms)
```

## For the given word print its hyponyms

```python
word = input()
word = wn.synsets(word)[0]
types = list(set([w for s in word.closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
print(sorted(types))
```
